```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score,ShuffleSplit,GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer,OneHotEncoder,OrdinalEncoder,StandardScaler,MinMaxScaler,RobustScaler
from sklearn.compose import ColumnTransformer
```

```python
data=pd.read_csv('train.csv')
data.head()
```

```python
#one hot encode 
data=pd.get_dummies(data)
```

```python
# there are so many variables--- try to pare it down? 
#can't even include the labels without basically killing the plot-- so hard to see what's what here

correlations=abs(data.corr())
heatmap=plt.pcolor(correlations)
plt.grid()
#plt.yticks(np.arange(0, len(correlations.index), 1), correlations.index)
#plt.xticks(np.arange(0, len(correlations.columns), 1), correlations.columns,rotation=90)
plt.colorbar(heatmap)
plt.show()
```

```python
#take top 50 correlated with sale price
top50 = correlations.nlargest(50, 'SalePrice')['SalePrice'].index
top50corrs=correlations[top50]
top50corrs=top50corrs.loc[top50]
```

```python
top50
#what do these mean? 
#SalePrice--what we want to predict
#OverallQual: Rates the overall material and finish of the house
#GrLivArea: Above grade (ground) living area square feet
#GarageCars: Size of garage in car capacity
#GarageArea: Size of garage in square feet---seems like this would be nearly perfectly correlated with garage cars
#TotalBsmtSF: Total square feet of basement area
#1stFlrSF: First Floor square feet
#FullBath: Full bathrooms above grade
#2ndFlrSF: Second floor square feet
#TotRmsAbvGrd: Total rooms above grade (does not include bathrooms)
#YearBuilt--self explanatory
#YearRemodAdd: Remodel date (same as construction date if no remodeling or additions)---need to see how often this actually differs from yearbuilt
#GarageYrBlt: Year garage was built
#MasVnrArea: Masonry veneer area in square feet
#Fireplaces: Number of fireplaces
#BsmtFinSF1: Type 1 finished square feet
#LotFrontage: Linear feet of street connected to property
#WoodDeckSF: Wood deck area in square feet
#OpenPorchSF: Open porch area in square feet
#HalfBath: Half baths above grade
```

```python
heatmap=plt.pcolor(top50corrs)
plt.grid()
plt.yticks(np.arange(0, len(top50), 1), top50)
plt.xticks(np.arange(0, len(top50), 1), top50,rotation=90)
plt.colorbar(heatmap)
plt.show()
#halfbath highly correlated with 2ndFlrSF, garagecars and garage area almost identical
```

```python
print(top50corrs)

#remove any numerical feature above .8 correlated with each other
# not sure about the categorical ones-- most are highly correlated to each other because if you are one you are not the other

#garage carrs and garage area highly correlated-- keep garage cars as slighlty better correlated with saleprice
#totalBsmtSF correlated with 1stFlrSF-- keep totalBsmtSF as slightly better correlated with saleprice
#grlivarea correlated with totalrmsabvgrd-keep grlivarea
#year buil highly correlated with garage year built-- keep yearbuilt

finalfeatures=[feature for feature in top50 if feature not in ['GarageArea','1stFlrSF','TotRmsAbvGrd','GarageYrBlt']]
```

```python
final_data=data[finalfeatures]
```

```python
final_data.info()
#LotFrontage, MsVnrArea missing values
```

```python
#standardize numerical values
train_labels=final_data['SalePrice']
train_examples=final_data.drop('SalePrice',axis=1)
numerical_vars=['GrLivArea','TotalBsmtSF','GarageCars','OverallQual','FullBath','YearBuilt','YearRemodAdd','MasVnrArea','Fireplaces','BsmtFinSF1','LotFrontage','WoodDeckSF','2ndFlrSF','OpenPorchSF','HalfBath']
categorical_vars=[var for var in train_examples.columns if var not in numerical_vars]
```

```python
#transform numerical_vars
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler())])
#used get_dummies instead bc easier, but not sure if same and may come back 
#to this instead
#cat_pipeline= Pipeline([
    #only one item, but make pipeline incase more are added later
    #('onehot',OneHotEncoder()),
#])

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, numerical_vars),
        #("cat", cat_pipeline, categorical_vars),
    ])

train_prepared=full_pipeline.fit_transform(train_examples)
```

```python
def rmse(scores):
    return np.sqrt(-np.mean(scores))
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor()
model.fit(train_prepared,train_labels)
scores = cross_val_score(model, train_prepared, train_labels,
                         scoring="neg_mean_squared_error", cv=5)

print(rmse(scores))
#wow it is horrible
```

```python
#regularized models--Ridge

from sklearn.linear_model import Ridge, ElasticNet, Lasso

model=Ridge()
model.fit(train_prepared,train_labels)
scores = cross_val_score(model, train_prepared, train_labels,
                         scoring="neg_mean_squared_error", cv=5)

print(rmse(scores))
```

```python
model=ElasticNet()
model.fit(train_prepared,train_labels)
scores = cross_val_score(model, train_prepared, train_labels,
                         scoring="neg_mean_squared_error", cv=5)

print(rmse(scores))
```

```python
model=Lasso()
model.fit(train_prepared,train_labels)
scores = cross_val_score(model, train_prepared, train_labels,
                         scoring="neg_mean_squared_error", cv=5)

print(rmse(scores))
```

```python
#so far ridge is best
from sklearn.svm import SVR
model=SVR()
model.fit(train_prepared,train_labels)
scores = cross_val_score(model, train_prepared, train_labels,
                         scoring="neg_mean_squared_error", cv=5)

print(rmse(scores))
```

```python
#SVR,Ridge, RF best with transformed labels, but accuracy is still quite low compared to leaderboard values
#let's try more thorough treatment of all variables

#get all features with missing data
data=pd.read_csv('train.csv')
data_null=data.isnull().sum()
null_labels=data_null[data_null>0].index
```

```python
null_labels
#what does each null value represent, what can we replace with
#lot frontage, MasVnrArea are replaced with median by pipeline
#from description-- null is no alley acess--replace string NA
#MasVnrType---no veneer, replase string NA
#BsmtQual--no basenment, replace with string NA
# Bsmt Cond--no basenment, replace with string NA
#BsmtExposure--no basenemnt, replace with string NA
#BsmtFinType1', 'BsmtFinType2', no basement, replace with string NA
#Electrical: no indication, replace with most common value
#FireplaceQu: no fireplace, replace with string NA
#GarageType: no garage, replace with string NA
#GarageYrBlt: no indication--probably no garage, replace with 0 since numerical
#GarageFinish, GarageQual,GarageCond: all mean no garage, replace string NA
#PoolQC: no pool, replace with string NA
#Fence: no fence--repalce with string NA
#MiscFeature: no other feature, replace with string NA
```

```python
#MMSSubClass is actually a category
data['MSSubClass']=data['MSSubClass'].apply(str)
data['MSSubClass']
```

```python
#new pipeline
none_vars=['Alley','MasVnrType','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']
other_vars=['GarageYrBlt']
num_vars=['YrSold','MoSold','MiscVal','GarageYrBlt','LotFrontage','MasVnrArea','LotArea','OverallQual','OverallCond','YearBuilt','YearRemodAdd','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtFullBath','BsmtHalfBath','FullBath','HalfBath','BedroomAbvGr','KitchenAbvGr','TotRmsAbvGrd','Fireplaces','GarageCars','GarageArea','WoodDeckSF','OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea']
ord_vars=['ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PavedDrive','Fence','PoolQC']
cat_vars=['SaleCondition','Electrical','MSSubClass','MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope','Neighborhood','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','Foundation','Heating','CentralAir','Functional','Alley','MasVnrType','BsmtExposure','BsmtFinType1','BsmtFinType2','GarageType','GarageFinish','MiscFeature','SaleType']


num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', StandardScaler())])

none_pipeline= Pipeline([
    ('imputer',SimpleImputer(strategy="constant",fill_value='None')),
    ('onehot',OneHotEncoder())
])

cat_pipeline= Pipeline([
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('onehot',OneHotEncoder())
])
other_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy="constant",fill_value=0))
])

ord_pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('intencoder',OrdinalEncoder())
])

full_pipeline = ColumnTransformer([
        ("none", none_pipeline, none_vars),
        ("other",other_pipeline,other_vars),
        ("num", num_pipeline, num_vars),
        ("cat", cat_pipeline, cat_vars),
        ("ord",ord_pipeline,ord_vars)
    ])

```

```python

train_examples=data.drop('SalePrice',axis=1)
train_labels=data['SalePrice']
train_labels_logs=np.log1p(train_labels)
train_prepared=full_pipeline.fit_transform(train_examples)
```

```python
model=ElasticNet()
model.fit(train_prepared,train_labels_logs)
scores = cross_val_score(model, train_prepared, train_labels_logs,
                         scoring="neg_mean_squared_error", cv=5)

rmse(scores)
```
